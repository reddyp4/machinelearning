```python
#Setup predictions
predictions = []
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs=()):
    predictions.append(model.predict(xs))
callbacks = myCallback()

#setup the numpy arrays
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

#Define the model

```python
SHAPE = [1] #Shape
LOSS = 'mean_squared_error' #average error

#Define the model
model = Sequential([Dense(units=1, input_shape=[1])])

#Compile model with optimizer and loss function
model.compile(optimizer='sgd', loss=LOSS)
```

#Include the call to fit the data

```python
#fit the model
model.fit(xs, ys, epochs=300, callbacks=[callbacks], verbose=2)
```

#Plot the predictions with epocs of 1,25,50,150,300.

```python
EPOCH_NUMBERS = [1,25,50,150,300] #Update to see other epochs
plt.plot(xs,ys,label = 'Ys')
for EPOCH in EPOCH_NUMBERS:
  plt.plot(xs,predictions[EPOCH-1],label = 'Epoch = ' + str(EPOCH))
plt.legend()
plt.show()
```
